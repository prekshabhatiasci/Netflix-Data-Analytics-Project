# Handle missing values:

```python
df["Director"] = df["Director"].fillna("Unknown")
df["Country"] = df["Country"].fillna("Unknown")

# Remove rows where Release_Date is missing
df.dropna(subset=["Release_Date"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Cast"] = df["Cast"].fillna("Unknown")
df["Rating"] = df["Rating"].fillna("Unknown")
```

```python
df.isnull().sum()
```

# Check for duplicate records:

```python
df.duplicated().sum()   # as a result we have two duplicate rows
```

# Remove duplication:

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# Converting data types:

```python
df["Release_Date"] = df["Release_Date"].str.strip()   # used to remove unwanted spaces between date, month, year
df["Release_Date"] = pd.to_datetime(df["Release_Date"])
df
```

```python
# extracting year and month :
df["Year Added"] = df["Release_Date"].dt.year
df["Month Added"] = df["Release_Date"].dt.month_name()
```
